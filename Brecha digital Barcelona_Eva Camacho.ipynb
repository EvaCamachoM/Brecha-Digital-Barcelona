Como último paso de la transformación, creo 4 medidas DAX que facilitarán la visibilidad de los datos en cuanto a proporciones y comparaciones entre los diferentes grupos encuestados:

1. <span style="color: grey; font-family: Verdana; font-size: 16px; font-weight: bold;">Total encuestados:</span> medida necesaria para calcular el porcentaje de la población sin internet con respecto al total de la muestra:

```
Total encuestados = COUNTROWS(Encuesta)
```

2. <span style="color: grey; font-family: Verdana; font-size: 16px; font-weight: bold;">Total de personas con acceso a internet en sus hogares:</span> servirá para cuantificar el número de personas que sí tienen acceso a internet y poder realizar un análisis más detallado sobre su distribución según segmento de población:

```
Con acceso a internet = COUNTROWS(FILTER(Encuesta,Encuesta[Acceso Internet]="Sí"))
```

3. <span style="color: grey; font-family: Verdana; font-size: 16px; font-weight: bold;">Total de personas sin acceso a internet:</span> servirá para cuantificar el número de personas que no tienen acceso a internet y poder identificar qué grupos de población pueden verse más afectados.

```
Sin acceso a internet = COUNTROWS(FILTER(Encuesta,Encuesta[Acceso Internet]="No"))
```

4. <span style="color: grey; font-family: Verdana; font-size: 16px; font-weight: bold;">Porcentaje de personas sin acceso con respecto al total de encuestados:</span> permitirá realizar análisis comparativos en diferentes contextos y saber si el acceso está distribuido equitativamente entre diferentes grupos:

```
% Personas sin acceso a internet = 
DIVIDE('Medidas (2)'[Sin acceso a internet],'Medidas (2)'[Total encuestados])
```

<span style="color: blue; font-family: Verdana; font-size: 18px; font-weight: bold;">RESULTADOS

<span style="color: blue; font-family: Verdana; font-size: 18px;">¿Qué proporción de personas hay que no disponen de acceso a internet en sus casas?

En la primera visualización de este análisis se observa la proporción de personas que no tienen acceso a internet en sus hogares en comparación con el total de encuestados. Este grupo representa aproximadamente un 8% del total de la muestra.

Aunque a primera vista este porcentaje puede parecer relativamente bajo, sigue siendo importante, sobre todo si consideramos el contexto actual, en el que gran parte de las actividades cotidianas, como la educación, el trabajo, las gestiones administrativas, el acceso a servicios de salud, etc dependen cada vez más de una conexión a internet. 

En un mundo cada vez más digitalizado, no disponer de internet en el hogar implica una potencial marginación de este 8% de la sociedad digital, afectando la capacidad de participar en procesos fundamentales como la banca online, la formación continua, y el acceso a información y servicios básicos.

<span style="color: blue; font-family: Verdana; font-size: 18px;">¿Qué relación existe entre la edad y el acceso a internet en los hogares?

Esta visualización muestra claramente cómo el porcentaje de personas sin acceso a internet en sus hogares aumenta conforme lo hace la edad. Esto confirma que el grupo más afectado por la brecha digital en Barcelona es el de las personas de 75 años o más con una tasa del 31%. En contraste, el segmento de la población menos afectado está compuesto por jóvenes de entre 16 y 24 años, cuya tasa es de apenas un 1%. Esto respalda la hipótesis de que la brecha digital tiene un importante componente generacional.

<span style="color: blue; font-family: Verdana; font-size: 18px;">¿Cómo afecta el nivel académico en el uso de las nuevas tecnologías?

En el siguiente gráfico se confirma que el acceso a internet está directamente relacionado con el nivel académico de los encuestados. <span style="color: blue; font-family: Verdana; font-size: 18px;">¿Existe brecha digital por diferencia de género y edad?

El resultado obtenido es el siguiente gráfico donde se puede observar que existe una brecha significativa de acceso a internet en las personas mayores de 75 años, sobre todo en el grupo femenino, donde el 35% de las mujeres no tiene acceso. Para los hombres en este mismo rango de edad, el 23% no tiene acceso. Este es el grupo con mayor proporción de personas sin conexión. Por contra, los grupos más jóvenes (16 a 44 años) tanto femeninos como masculinos tienen un acceso casi total a internet, con muy bajos porcentajes de personas sin acceso.

En resumen, la brecha digital se agrava en personas mayores, especialmente mujeres de más de 65 años, mientras que los grupos jóvenes tienen un acceso casi universal.

<span style="color: blue; font-family: Verdana; font-size: 18px; font-weight: bold;">CONCLUSIONES

Como conclusión final, destacaría la brecha digital que existe en algunos grupos encuestados:

* Mayores de 65 años: sin duda es el grupo más afectado por esta brecha. El motivo podría estar asociado a una combinación de factores, como la falta de habilidades digitales, la resistencia al cambio, las limitaciones económicas, infraestructura deficientes y/o las preferencias personales. Tratar estos aspectos es clave para cerrar la brecha y asegurar que todas las generaciones puedan beneficiarse de las oportunidades que ofrece internet.

* Otro grupo afectado por la brecha son las mujeres, especialmente dentro del grupo de personas mayores de 65 años. Esto podría ser debido a factores socioeconómicos y culturales que han afectado a las mujeres a lo largo de sus vidas. Es necesario implementar políticas y programas específicos que tengan en cuenta tanto la edad como el género para poder reducir esta desigualdad.

* El nivel académico tiene también un impacto directo en la brecha digital, ya que las personas con menos formación están en mayor riesgo de quedar excluidas de la sociedad digital. Esto refuerza la necesidad de implementar programas de alfabetización digital y políticas que faciliten el acceso a la tecnología para aquellos con menores niveles educativos con el fin de cerrar esta brecha y fomentar una mayor inclusión.
